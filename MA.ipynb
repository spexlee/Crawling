```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
con = sqlite3.connect("C:/Users/dpqls/PycharmProjects/stock_analysis/db/stock_db(day)_merge.db")
```

```python
df = pd.read_sql("SELECT day, cur_pr, high_pr, low_pr,clo_pr, acc_vol, com_buy_vol FROM A005930", con, index_col=None)
```

```python
df = df.rename(
    {
      'cur_pr': 'open'
    , 'high_pr':'high'
    , 'low_pr':'low'
    , 'clo_pr':'close'
    , 'acc_vol':'volume'
    , 'com_buy_vol':'comvol'
    }, axis='columns'
)
```

```python
ma5 = df['close'].rolling(5).mean()
ma10 = df['close'].rolling(10).mean()
ma20 = df['close'].rolling(20).mean()
ma60 = df['close'].rolling(60).mean()

df['ma5'] = ma5
df['ma10'] = ma10
df['ma20'] = ma20
df['ma60'] = ma60
```
